```python
def unique(Y): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in Y: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    # print list 
#     for x in unique_list: 
#         print (x)
    return unique_list

Y = df.Source;
A = unique(Y);
print(A)
lA = len(A)
print(lA)

Z = df.Target;
B = unique(Z);
lB = len(B)
print(lB)

V = unique(df.Site_target);



df[df.values  == "AcuteMedicalUnit.EAU"]
```
