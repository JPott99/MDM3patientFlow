```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
allData = pd.read_csv("journal.pone.0185912.s003.csv")
allData.head()
```

```python
allData['Transfer_Name'] = allData.Source + allData.Target
allData['WeekFromStart'] = (allData.Year-2014)*52 + allData.Week-47

listOfWeeks = np.unique(allData.WeekFromStart.values)

allData.head(20)  
```

```python
Data14 = allData[allData['Year'] == 2014]
np.unique(Data14.Week.values)
```

```python
transferVariation = pd.DataFrame(columns = ['Mean', 'StandardDev', 'TransferName'])
for transfer in np.unique(allData.Transfer_Name.values):
    weeklyTransfers = allData[allData['Transfer_Name'] == transfer]
    weeklyTransfers = weeklyTransfers[['WeekFromStart', 'Transfer_Name', 'Transfers']]
    weeksZero = np.setdiff1d(listOfWeeks, np.unique(weeklyTransfers.WeekFromStart.values))
    
    df = pd.DataFrame(weeksZero)
    df.columns = ['WeekFromStart']
    df['Transfer_Name'] = transfer
    df['Transfers'] = 0
    #print(df.shape,weeklyTransfers.shape)
    weeklyTransfers = weeklyTransfers.append(df)
    weeklyTransfers.sort_values(by=['WeekFromStart'], inplace = True)
    #print(weeklyTransfers.head())
    
    if weeklyTransfers.Transfers.sum() > 5:
        transferStd = weeklyTransfers.Transfers.std()
        transferMean = weeklyTransfers.Transfers.mean()
        newRow = {'Mean':transferMean, 'StandardDev':transferStd, 'TransferName':transfer}
        transferVariation = transferVariation.append(newRow, ignore_index = True)
        #print(transferStd,transferMean,transfer)
        
  
```
